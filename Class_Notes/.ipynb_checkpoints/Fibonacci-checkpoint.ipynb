```python
def Fibonacci(n):
    array = []
    def recursive_fibonacci(n):
        if n < array.len():
            return array[n]
        if n == 0:
            res = 0
        elif n == 1:
            res = 1
        else:
            res = recursive_fibonacci(n - 1) + recursive_fibonacci(n - 2)
            
        array.push(res)
        return res
    return recursive_fibonacci(n)

Fibonacci(2)
```
