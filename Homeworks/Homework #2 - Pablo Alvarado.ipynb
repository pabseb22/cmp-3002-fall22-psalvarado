```python
class Node:
    """
    Implementation of a node
    """
    def __init__(self, val=None):
        self.val = val
        self.next_node = None
    
    def set_next_node(self, next_node):
        self.next_node = next_node

import ctypes
class Stack(object):
    """
    Implementation of the stack data structure
    """

    def __init__(self, n):
        self.l = 0
        self.n = n
        self.head_node = Node()

class Stack(Stack):
    def push(self, new_node):
        """
        Add new item to the stack
        """
        if self.l == self.n:
            raise ValueError("no more capacity")
        
        c = self.head_node
        self.head_node = (new_node)
        self.head_node.next_node = c
        self.l += 1
        
    def pop(self):
        """
        Remove an element from the stack
        """
        if not self.l:
            raise('stack is empty')
            
        c = self.head_node
        next = self.head_node.next_node
        self.head_node = None
        self.head_node = next
        self.l -= 1
        return c
    
    def top(self):
        """
        Show the top element of the stack
        """
        return self.head_node.val
    
    def full(self):
        """
        Is the stack full?
        """
        return self.l == self.n


    def empty(self):
        """
        Is the stack empty?
        """
        return self.l == 0


    def size(self):
        """
        Return size of the stack
        """
        return self.l

# Implementation Test:
    
stack = Stack(3)
stack.push(Node("1"))
stack.push(Node("2"))
stack.push(Node("3"))
print(stack.top())
print(stack.full())
stack.pop()
stack.pop()
stack.pop()
print(stack.size())
```

```python
class PriorityQueue(object):
    """
    Implementation of the queue data structure
    """

    def __init__(self, n):
        self.l = 0
        self.n = n
        self.queue = self._create_queue(self.n) 
    
    def _create_queue(self, n):
        """
        Creates a new stack of capacity n
        """
        return (n * ctypes.py_object)()
    
    def empty(self):
        return len(self.queue) == 0
    
    def enqueue(self, item):
        """
        Add new item to the queue
        """
        if self.l == self.n:
            raise ValueError("no more capacity")
        self.queue[self.l] = item
        self.l += 1
        
    def dequeue(self):
        max_val = 0
        for i in range(self.l):
            if self.queue[i] > self.queue[max_val]:
                max_val = i
        temp = self.queue[max_val]
        self.queue[max_val] = None
        return temp

q = PriorityQueue(10)

q.enqueue((1,2))
q.enqueue((2,4))
q.enqueue((0,1))
q.enqueue((9,43))
q.enqueue((1,21))

print(q.queue[0:q.l])
print(sorted(q.queue[0:q.l]))

print(q.dequeue())

print(q.queue[0:q.l])
```

### Complexity and Runtime Analysis

The enqueue method has a O(1) since it is only adding a value to the queue. However, the dequeue method requires to revise the whole queue in order to determine its priority, which is why it has a O(n).
