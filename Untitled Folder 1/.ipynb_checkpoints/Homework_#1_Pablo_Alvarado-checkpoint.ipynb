### 1. Assume we implemented an algorithm that has complexity approximately $n log(n)$. How much faster is this implementation compared to one of complexity $n^2$?

First, we could plot the relation between $n^2$ and $n log(n)$, so that first we could have a direct view of how much faster is $n log(n)$.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

def plot_comparison(range, step):
    r = pd.DataFrame()
    r['n'] = np.arange(1,range,step)
    r['nlog(n)'] = r['n'] * np.log2(r['n'])
    r['n^2'] = np.power(r['n'],2)
    r.plot(x='n',figsize=(7,3))
    plt.xlabel('n')
    plt.grid()

plot_comparison(10,1)
plot_comparison(100,10)
plot_comparison(1000,100)
````

As we can clearly see, $n^2$ grows a lot faster than $n log(n)$. Showing us that a $n log(n)$ complexity is prefereable.   However, in order to determine how much faster it is, we need to compare it by:

First, we need to compare and validate that both functions return de the same outputs:

````python
def sum1(n):
    total = 0
    for i in range(1, n+1):
        total += i
    return total

def sum2(n): 
    total = n*(n+1)//2
    return total

def validate(values):
    print("{:<12} {:<12}".format("Suma 1","Suma 2"))
    print("{:<12} {:<12}".format("------","------"))
    for i in values:
        print("{:<12} {:<12}".format(sum1(i), sum2(i)))
        
        
A = [5, 8, 103, 527]

validate(A)
        
````

Now, we can continue with the comparison of the runtime of this two funcitons. The function used to estimate the runtime is the following: 

````python

from time import time
  
  
def estimate_runtime(function):
    def wrap_f(*args, **kwargs):
        t1 = time()
        result = function(*args, **kwargs)
        t2 = time()
        print(f'Function {function.__name__!r} executed in {(t2-t1):.4f}s')
        return result
    return wrap_f
  
#Which allows us to define the two functions as:
    
@estimate_runtime
def sum1(n):
    total = 0
    for i in range(1, n+1):
        total += i
    return total

@estimate_runtime
def sum2(n): 
    total = n*(n+1)//2
    return total

def validate_runtime(values):
    for i in values:
        print("Suma 1: {:<12} Suma 2: {:<12}".format(sum1(i), sum2(i)))
        print("")
    

A = [1, 10, 100, 1000, 10000, 100000, 1000000, 10000000, 1000000000, 1000000000]

validate_runtime(A)
````
