### A capability framework for product engineering teams - Juan Pablo @buritica

As we all know, programming is mostly an individual activity, while building software products is a social one. Collaboration is an emergent property of teams and team effectiveness doesn't fall solely on management. If we can have competency frameworks for individuals, why can't we establish one for teams? This is the main idea that Juan Pablo wants to transmit inside his conferences, dividing it in key factors that validates his ideas.

Capabilities help us establish expectations as outcomes, and ownership over path enables autonomy. In this context a framework allows the team to focus on results. Therefore, Juan Pablo established a framework that describes what a capable product engineering team must be able to do:
1. Navigate ambiguity.
•	Find valuable work instead of idling waiting for direction.
•	Find the right places and people to ask questions.
•	Ask clarifying questions and seek aligment.
2. Set Goals:
•	Define outcomes that align with overall strategy.
•	Establish goals that are measurable and achievable.
•	Communicate engineering approaches effectively through business outcomes.
3. Ask for help:
•	Raise their hands with concrete requests for help.
•	Surface needs without waiting for meetings or expiring deadlines.
•	Offer alternative resolutions and negotiate if necessary.
4. Broadcast state:
•	Establish or use distribution channels for status reporting.
•	Communicate "at risk" or "off track" status proactively, ideally with resolution plans.
•	Minimize surprises on dependent teams or leadership.
5. Negotiate dependencies:
•	Proactively track and align dependencies across teams.
•	Communicate changes and priority shifts.
•	Facilitate contributions across codebases as needed to reduce coupling.
6. Escalate appropriately:
•	Start by communicating directly with cross-functional peers.
•	Document the conflict, possible resolutions and existing discussions.
•	Escalate through established channels, leaning on leadership to tie-break and take responsibility for outcomes.
7. Manage it's capacity:
•	Understand its distribution across maintain, support & feature building.
•	Balance accordingly, accelerating tempo without overworking.
•	Consistently deliver on 80%+ of planned work.
8. onboard team members:
•	Enable new (or visiting) team-members to feel comfortable and productive quickly.
•	Establish self-service environment with support channels.
•	Encourage and reward team members who continuously improve onboarding for future members.
9. Understand it's users:
•	Has collective familiarity over needs. 
•	Prioritize across perceived and demonstrated friction.
•	Has established user feedback channels.
10. Understand it's levers:
•	Connect organizational business goals to day to day work.
•	Invest appropriately in impactful work.
•	Translate "technical debt" to business impact. 
11. Tell it's story:
•	Constantly sharing the narrative that contextualizes work.
•	Influence peer teams to use similar language and connect goals.
•	Persistent "political campaign"
12. Define working agreements:
•	Discuss and negotiate ways of working that serve team-members.
•	Iterate on established agreements as team evolves. 
•	Adopt sensible organizational defaults to facilitate predictable cadence.
13. Adopt an operational cadence:
•	Converge on organizational rhythm. 
•	Facilitate team observability.
•	Lean on predictable tempo to facilitate focus
14. Anticipate problems:
•	Use monitoring, alerting and other tools to observe and react.
•	Instrument perceived friction to validate effort investments.
•	Plan and document releases effectively. 
•	Awareness over important organizational events. 
15. Respond to incidents:
•	Establish, documented, on-call and paging procedures. 
•	Create and curate run-books reducing ambiguity and anxiety. 
•	Confidently escalate and page as needed.
•	Communicate response state clearly and in a timely manner. 
16. Improve on it's processes:
•	Continuously evaluates what's working well and what can improve.
•	Improvement isn't restricted or solely directed by management.
•	Empowers team to develop tooling that eliminates toil.
17. Tackle increasing complexity:
•	Consistently reduce risk of increasingly larger projects. 
•	Shift from following to driving strategy.
•	Operating and executing on expanding or simultaneous horizons.

Despite it presenting us a broad description of a framework that will improve a team effectivenes, the main objective of this conference was to highlight the need of understanding the complexity in team relations, and how each member and leader needs to work in a define and coherent environment. Which is why, Juan Pablo encourages us to consider the main topics of a a capable team must be capable of.

### Towards a responsible artificial intelligence - Carla Mamani y Luis Beltrán

Artificial intelligence nowadays has become a recurrent topic and idea that not only demonstrates the great power that software development has, but, encourages each of us by setting realistic objectives that will then generate huge impacts in our society. However, as we all know, with great power comes great responsability. And this idea is exactly what this conference wants to transmit to us, as artificial intelligence has established as something "normal" in our lives. 

The conference starts by highlighting how the bast majority of people do not trust artificial intelligence or even think that it is responsible. After this, the principles of a responsible AI were established: 
1. Fairness: Looks for the absence of negative impacts on groups based on sensitive characteristics such as gender, age, origin and others. However, in order to accomplish this, it is required to create parity restrictions, that will not allow to make relations for negative impacts.
2. Reliability and security: We have to have robust monitoring and follow-up procedures to modeling so that the performance of the model is optimal. Also, sensitive information of the users must be taken into account in order to avoid any serious security issue.
3. Privacy: The models must not enter information that it was not previously allowed, in order to mantain a certain level of privacy.
4. Inclusion: We have to design an AI that is undiscriminatory but also we have the ticket to account the fact that the world around us is very diverse we have people with disabilities or thank you basically AI systems need to be inclusive and they have to be in tune with the needs of our diverse ecosystem when artificial intelligence systems.
5. Transparency and responsability: There must be packages that contribute to the transparency and explainability of a model, such us: Interpret-community, InterpretML, Fairlearn. Also, responsability of the owners of each model and what it causes is essential.

Then, we also have a "Responsible AI Toolbox" that allows to analyze our models by having available:
1. Responsible AI Dashboard
2. Fairness Dashboard
3. Error Analysis Dashboard
4. Interpretability Dashboard

In conclusion, there are a lot of things and aspects to consider if we want to be responsible with AI, and despite the conference gave us a lot of detailed information on its importance and how to apply, we also have some general recommendations such us:
1. Clarify what the smart system is going to do
2. Clarify system performance
3. Display relevant contextual information
