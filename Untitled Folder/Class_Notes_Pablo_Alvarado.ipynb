#### What is an <span style="color:blue">Algorithm</span>?

- An algorithm can be defined as a procedure for solving a mathematical problem in a finite number of steps.
- Algorithms ussually involve the repetition of steps or operations.
- A more general definition involves that algorithms are a step-by-step procedure for solving a problem or accomplish an objetctive.
- In computational terms an algorithm represents a mathematical abstraction of a computer program, or a computational process to solve a problem. By the model of: input -> algorithm -> output

Some famous analogies of algorithms are:
- program -> algorithm
- language -> pseudocode
- computer -> model of computation


#### What is a <span style="color:blue">Model of Computation</span>?

A model of computation is what defines or specifies the operations that you can do in an algorithm.How much does these operations cost (time, memory, etc). Then,we look how much each operation costs and add them up to get the total cost of the algorithm. 

Models of computation can be thought as styles of programming (ex: object oriented programming). Where in these course we'll cover two models of computation: <span style="color:green;">random access machine</span> and <span style="color:green;">pointer machine</span>


#### <span style="color:green;">Random Access Machine (RAM)</span>

Represents a single processor (operations are executed one after the other). It could be modeled by a big array and it has constant number of registers. This is a realistic and powerful model as it allows us to compare algorithms on a performance basis. Where we have to consider that:

- Each operation takes 1 time steps (write or read)
- Needs to know where the words are to access them
- A word is an x number of bits


#### <span style="color:green;">Pointer Machine</span>

It represents a Simple version of object oriented programming (OOP), where exists dynamically allocated objects (create and destroy), objects have a constant number of fields and a field is either a word or a pointer.

In this case pointer either points to an object or is null (none in Python). To analyze it, we need to consider that:

- Adding values takes 1 time steps
- Following a pointer takes 1 time steps

This is a weaker model than RAM since you can build this model with RAM as it can be also transfered to operations executed one after the other.

#### Python model

Python offers both perspectives (RAM and pointer machine)

1. Python lists are arrays (RAM)
```
L[i] = L[i] + 1
```

2. Objects with a constant number of attributes (pointer machine)

```
x.val
x.next()
```

Both operations take constant time (finite number of time steps)

#### How to time our code?


````python
import time 

def sum1_time(n):
    start = 1000000*time.time()
    total = 0
    for i in range(1,n+1):
        total += i
    return 1000000*time.time() - start


````

We neede to Implement this using Python decorators.

Decorators:are a very powerful and useful tool in Python since it allows programmers to modify the behaviour of a function or class. Decorators allow us to wrap another function in order to extend the behaviour of the wrapped function, without permanently modifying it.

#### Math Models

Total runtime can be defined as the sum of the cost of the basic operations of the programs multiply by their frequency. For example:
```
def sum1(n):
    total = 0
    for i in range(1, n+1):
        total += i
    return total
```


|Operation|Cost|Frequency|
|---------|----|---------|
| var declaration | $c_1$ | 2 |
| array access | $c_2$ | $n$ |
| increment | $c_3$ | $n$ |

> $T(n) = 2c_1 + (c_2 + c_3)n$


On the other hand, we can also simplify this process by:

1. Just counting the most expensive ones
    - or the one executed the most frequently
2. Estimate the runtime as a function of $n$
    - ignore lower order terms
    - the limit to infinity is 1
    
> $T(n) = 2c_1 + (c_2 + c_3)n$

> $T(n) \approx (c_2 + c_3)n$
 
 
 
 
### Insertion sort

- Input: A sequence of $n$ numbers $(a_1, a_2, \dots, a_n)$
- Output: A permutation $(a'_1, a'_2, \dots, a'_n)$ of the input sequence such that $a'_1 \leq a'_2 \leq \dots \leq a'_n$

### Asymptotic notation

- Domain is the natural numbers $\{0, 1, 2, \dots\}$
- Describe the worst-case of $T(n)$

Types of analyses

1. **Best case:** lower bound on cost, based on $\Omega$ notation
    - easiest input
    - provides a goal for all inputs
    
2. **Worst case:** upper bound on cost, based on $O$ notation
    - most difficult input
    - provides a guarantee for all inputs
    
3. **Average case:** expected cost for a random input
    - need to model for a random input
    - provides a way to predict performance

#### $\Theta-notation$

> $\Theta(g(n)) = \{f(n) : \exists c_1, c_2, n_0 > 0, \text{such that } 0 \leq c_1g(n) \leq f(n) \leq c_2g(n), \forall n \geq n_0 \}$

- tight bound
- we need two constants that represent an upper bound and a lower bound
- Example: for loop to calculate sum of 1 to n is $\Theta(n)$

#### $O-notation$

> $O(g(n)) = \{f(n) : \exists c, n_0 > 0 \text{ such that } 0 \leq f(n) \leq cg(n), \forall n \geq n_0 \}$

- upper bound
- Example: insertion sort is $O(n^2)$

#### $\Omega-notation$

> $\Omega(g(n)) = \{f(n) : \exists c, n_0 > 0 \text{ such that } 0 \leq cg(n) \leq f(n), \forall n \geq n_0 \}$

- lower bound
- Example: insertion sort is $\Omega(n)$

** Although is the standard, focusing on the worst case might not be the best approach
